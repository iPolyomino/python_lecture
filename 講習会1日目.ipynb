### 標準偏差の求め方

$$
s = \sqrt{s^2} = \sqrt{\frac{1}{n}\sum_{n = 1}^{n}(x_i - \overline{x}) ^ 2}
$$

````python
monk_fish_team = [158, 157, 163, 157, 145]

total = sum(monk_fish_team)
length = len(monk_fish_team)
mean = total/length
variance = 0

for height in monk_fish_team:
    variance += (height - mean)**2
    
variance = variance/length
variance**0.5
````

````python
volleyball_team = [143, 167, 170, 165]

total2 = sum(volleyball_team)
length2 = len(volleyball_team)
mean2 = total2/length2
variance2 = 0

for height in volleyball_team:
    variance2 += (height - mean2)**2
    
variance2 = variance2/length2
variance2**0.5
````

### 12引く9(偶数引く奇数)

- 2人対戦ゲーム
- お互いが1回ずつ6つサイコロを振る
        
        - 点数が大き方が勝ち
        - サイコロの目が偶数 → その目を加算
        - サイコロの目が奇数　→ その目を減算
        
- 例　→ Bさんの勝ち  
Aさん 1,2,3,4,5,6 → 2+4+6-1-3-5 = 3点  
Bさん 2,2,3,6,6,6 → 2+2+6+6+6-3 = 19点

> 乱数の用い方  
> p.386
>
> 1~6の乱数を作るには，以下のように書くといい．
> 
> ```
import random
random.randint(1, 6)
```

#### まずは1人のときを考える．

````python
import random

score = 0

for _ in range(6):
    dice = random.randint(1,6)
    if dice % 2 == 0:
        score += dice
    else:
        score -= dice

print(score)
````

#### 2人でスコアを競う

````python
import random

a_score = 0
b_score = 0

for _ in range(6):
    dice = random.randint(1,6)
    if dice % 2 == 0:
        a_score += dice
    else:
        a_score -= dice
        
for _ in range(6):
    dice = random.randint(1,6)
    if dice % 2 == 0:
        b_score += dice
    else:
        b_score -= dice

if a_score > b_score:
    print("aさんの勝ち")
elif a_score < b_score:
    print("bさんの勝ち")
else:
    print("引き分け")
````

### バーブディ

- 2人対戦ゲーム
- 交互に2つサイコロを振る．
    - 6-6, 5-6, 5-5, 4-4 が出たら勝ち
    - 1-1, 1-2, 2-2, 3-3が出たら負け

````python
import random

turn = 0
result = ""
while (True):
    turn += 1
    dice1 = random.randint(1, 6)
    dice2 = random.randint(1, 6)
    if (dice1 == 6 and dice2 == 6) or (dice1 == 5 and dice2 == 6) or (dice1 == 6 and dice2 == 5) or (dice1 == 5 and dice2 == 5) or (dice1 == 4 and dice2 == 4):
        result = "勝ち"
        break
    elif (dice1 == 1 and dice2 == 1) or (dice1 == 1 and dice2 == 2) or (dice1 == 2 and dice2 == 1) or (dice1 == 2 and dice2 == 2) or (dice1 == 3 and dice2 == 3):
        result = "負け"
        break

if turn % 2 == 0:
    print("Aさんの"+result)
else:
    print("Bさんの"+result)
````
